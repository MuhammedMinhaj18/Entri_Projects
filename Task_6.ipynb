##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:



image.png

````python
import matplotlib.pyplot as plt
x1,y1,y2 = [10,20,30],[20,40,10],[40,10,30]

plt.plot(x1,y1, 'b', linewidth = 3, label='line1- width-3')
plt.plot(x1,y2, 'r', linewidth = 5, label='line2- width-5')
plt.title('Two or more lines of different width and color, with legends')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.legend()
plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2,17.6,8.8,8,7.7,6.7]
plt.bar(x,y,color='Blue',edgecolor='Black',align='edge')
plt.title("Popularity of Programming Language \n"+"Worldwide, Oct2017 compared a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.minorticks_on()
plt.xlim(0,len(x))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='Red')
plt.grid(which='minor',linestyle=':',linewidth='0.5',color='Black')
plt.show()
````

````python
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
x=df.month_number
plt.plot(x,df.facecream,'-o', label='Face cream Sales Data',linewidth=3)
plt.plot(x,df.facewash,'-o', label='Face wash Sales Data',linewidth=3)
plt.plot(x,df.toothpaste,'-o', label='Toothpaste Sales Data',linewidth=3)
plt.plot(x,df.bathingsoap,'-o', label='Bathing Soap Sales Data',linewidth=3)
plt.plot(x,df.shampoo,'-o', label='Shampoo Sales Data',linewidth=3)
plt.plot(x,df.moisturizer,'-o', label='Moisturizer Sales Data',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend()
plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('company_sales_data.csv')
sales_data = (df.facecream.sum(),df.facewash.sum(),df.toothpaste.sum(),df.bathingsoap.sum(),df.shampoo.sum(),df.moisturizer.sum())
mylabel = ["FaceCream","FaceWash","Toothpaste","BathingSoap","Shampoo","Moisturizer"]
# plt.pie(sales_data,labels=["FaceCream","FaceWash","Toothpaste","BathingSoap","Shampoo","Moisturizer"],autopct='%0.1f%%')
plt.pie(sales_data,labels=mylabel,autopct='%0.1f%%')
plt.title("Sales Data")
plt.legend(loc='lower right')
plt.axis('equal')
plt.show()
````
