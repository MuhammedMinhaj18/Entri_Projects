10. Write a Python a program to find the union and intersection of two lists.


````python
l9=['ab',1,8,6,'ce','abc',18]
l10=[18,36,'acd','ce',45,8]

s1=set(l9)
s2=set(l10)

x1=s1.union(s2)
x2=s1.intersection(s2)

print('the union of the two lists = ',list(x1))
print('the intersection of the two lists = ',list(x2))
````

11.Write a Python script to sort (ascending and descending) a dictionary by value


````python
birthyears = {
    'Eshan': 2021,
    'Amin' : 2019,
    'Adarsh': 1996,
    'Aljo' : 1999,
    'Aiswarya': 1995,
    'shahanas': 1994
}
sort = sorted(birthyears.items(), key=lambda x:x[1])
sortdict=dict(sort)
print(sortdict)
````
